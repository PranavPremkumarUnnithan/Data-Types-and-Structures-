1. What are data structures, and why are they important?

  - A data structure is a way to organize and store data so it can be used efficiently.
  - They make tasks like searching, sorting, and updating data faster.
  - They are used in apps, websites, games, and other programs to handle data smartly.
2.  Explain the difference between mutable and immutable data types with examples?

  - Mutable:   Can be changed after creation.
   -Examples:  Dictionary,List
   -Immutable: Cannot be changed after creation
   -Examples:  String,Tuple

4. Describe how dictionaries store data?

  -Dictionaries store data in key-value pairs.
  -Keys are converted into a special number called a hash.
  -This number decides where the data is stored in memory.

5. Why might you use a set instead of a list in Python?

  - A set keeps only unique items.
  - Checking if an item is in a set is quicker than in a list.
  - Easy to find common or different items between sets.

6. What is a string in Python, and how is it different from a list?

  -A string is a sequence of characters, like "hello" or "123".
  -Strings are immutable, meaning once created, you can’t change the characters inside them.
  -Mutability,Syntax,Length

7. How do tuples ensure data integrity in Python?

 - Tuples ensure data integrity by encrypting the data inside them, making it secure and unchangeable.

8. What is a hash table, and how does it relate to dictionaries in Python?

  -A hash table is a data structure that stores data as key-value pairs.
  -It uses a hash function to convert keys into unique numbers.
  -Dictionaries in Python are built using hash tables.
  -Keys are hashed to find their place in memory.

9. Can lists contain different data types in Python?

  - Lists can hold numbers, strings, booleans, and even other lists or objects, all in one list.

13. How do sets handle duplicate values in Python?
  - Sets automatically remove duplicates.

   - For Lists:Checks if a value is in the list.
